```python


import pandas as pd

def input_age(cols):
    Age = cols.iloc[0]  # Use .iloc for positional indexing
    Pclass = cols.iloc[1]

    if pd.isnull(Age):  # Check if Age is missing
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age



```

```python

titanic['Age'] = titanic[['Age', 'Pclass']].apply(input_age, axis=1)
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False,cmap="viridis")
plt.show()
```
